```python
# Function to validate recharge plan
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        return False
    elif amount not in valid_plans:
        return False
    else:
        return True


# Main program with retry using while loop
while True:
    try:
        recharge_amount = int(input("Enter recharge amount: ₹"))

        if validate_recharge(recharge_amount):
            print("Recharge successful with plan ₹", recharge_amount)
            break
        else:
            print("Invalid recharge amount or plan. Please try again.")

    except ValueError:
        print("Please enter a valid number.")
```

```python
# Function to check stock levels
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(product, ": Reorder Alert (Stock =", stock, ")")
        else:
            print(product, ": Stock OK (Stock =", stock, ")")


# Dictionary of products and stock quantities
inventory = {
    "Rice": 10,
    "Wheat": 25,
    "Sugar": 12,
    "Oil": 30,
    "Salt": 8
}

# Call the function
check_inventory(inventory)
```

```python
# Function to process student marks
def student_result(marks):
    total = 0

    # Loop to calculate total marks
    for m in marks:
        total += m

    average = total / len(marks)

    # Check pass or fail
    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"

    return average, result


# Example list of marks
marks_list = [60, 55, 40, 70, 50]

# Call function
avg, res = student_result(marks_list)

print("Average Marks:", avg)
print("Result:", res)
```

```python
# Function to calculate cab fare
def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12

    fare = base_fare + (distance * per_km)

    if peak.lower() == "yes":
        fare = fare + (fare * 0.25)   # 25% extra

    return fare


# Retry option using while loop
while True:
    distance = float(input("Enter distance in km: "))
    peak_hour = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(distance, peak_hour)

    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

```python
# Function to check attendance eligibility
def check_attendance(attendance):
    present_days = 0

    # Loop to count present days
    for day in attendance:
        if day == "P":
            present_days += 1

    # Calculate attendance percentage
    percentage = (present_days / len(attendance)) * 100

    # Check eligibility
    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"


# Example attendance list
attendance_list = ["P", "A", "P", "P", "A", "P", "P", "P"]

# Call the function
percent, result = check_attendance(attendance_list)

print("Attendance Percentage:", percent)
print("Status:", result)
```

```python
# Function to check password strength
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Loop through password characters
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    # Check conditions
    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


# Input password
pwd = input("Enter password: ")

# Call function
result = check_password(pwd)

print(result)
```
