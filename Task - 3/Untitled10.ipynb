```python
# Dictionary of employees and their performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find the highest score among all employees
max_score = max(employees.values())

# Step 2: Find all employees who have the highest score (handles ties)
top_performers = [
    name for name, score in employees.items()
    if score == max_score
]

# Step 3: Print the result
print("Top score:", max_score)
print("Top performer(s):", top_performers)
```

```python
# Input search query
query = "Buy mobile phone buy phone online"

# Step 1: Convert the input to lowercase
query = query.lower()

# Step 2: Remove common punctuation
# (You can add more punctuation if needed)
punctuation = ".,!?;:\"'()[]{}"
for char in punctuation:
    query = query.replace(char, "")

# Step 3: Split the sentence into words (keywords)
words = query.split()

# Step 4: Count frequency of each keyword
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Step 5: Filter and display only keywords searched more than once
result = {word: count for word, count in frequency.items() if count > 1}

# Step 6: Print the result
print(result)
```

```python
# List of sensor readings where index = hour, value = reading
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Step 1: Create an empty list to store valid (hour, value) pairs
valid_readings = []

# Step 2: Loop through the list with index using enumerate()
for hour, value in enumerate(sensor_readings):

    # Step 3: Check if the reading is even (valid)
    if value % 2 == 0:

        # Step 4: Store (hour_index, reading_value) as a tuple
        valid_readings.append((hour, value))

# Step 5: Print the result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
from collections import Counter

# Input list of employee emails
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Step 1: Extract domains from each email
domains = []
for email in emails:
    domain = email.split("@")[1]   # Get the part after '@'
    domains.append(domain)


# Step 2: Count how many times each domain appears
domain_counts = Counter(domains)

# Step 3: Find total number of emails
total_emails = len(emails)

# Step 4: Calculate and print percentage usage of each domain
for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Input list of daily sales
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate the average daily sales
average_sales = sum(sales) / len(sales)

# Step 2: Define the spike threshold (30% above average)
threshold = average_sales * 1.3

# Step 3: Loop through sales with day number (starting from 1)
for day, value in enumerate(sales, start=1):

    # Step 4: Check if sales are more than 30% above average
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
from collections import Counter

# Input list of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Step 1: Count how many times each user ID appears
counts = Counter(user_ids)

# Step 2: Loop through the counts and print only duplicates
for user_id, count in counts.items():
    if count > 1:
        print(f"{user_id} → {count} times")
```
