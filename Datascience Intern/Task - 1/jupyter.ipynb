# **Task - 1**

1.In Natural Language Processing, "Love" and "love" are treated as different tokens when the text is case-sensitive. This means the model may consider them as two separate words. However, in most NLP pipelines, text normalization (lowercasing) is applied. After converting to lowercase, both words become the same
"Love" → "love"
"love" → "love"
This helps in reducing redundancy and improving model performance.

2.If stopwords are not removed:
The dataset becomes noisy and large
Models may focus on irrelevant words (like "is", "the", "and")
Training time increases
Accuracy may decrease because important words get diluted

3.Scenario 1: Sentiment Analysis
Sentence:"I am not happy"
If "not" is removed →"I am happy"(meaning completely changes)
Scenario 2: Question Answering Systems
Sentence:"What is the time?"
If "what" is removed →"is time?"(question loses meaning)

4.Stemming:
i.   Removes suffixes using simple rules
ii.  May produce incorrect words
iii. Faster but less accurate
iv.  Example: "running" → "runn"

Lemmatization:
i.   Uses vocabulary and grammar rules
ii.  Produces meaningful root words
iii. Slower but more accurate
iv.  Example: "running" → "run"

# **Task - 2**

```python
import re

def preprocess_text(text):
    if not text or not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Lowercase
    text = text.lower()

    # Remove URLs & emails
    text = re.sub(r"http\S+|www\S+|@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Remove repeated characters (soooo → soo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove special characters/emojis
    text = re.sub(r"[^a-z\s]", "", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenize
    tokens = text.split()

    # Remove short tokens (≤2) except important ones
    important_words = {"no", "not"}
    tokens = [word for word in tokens if len(word) > 2 or word in important_words]

    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence
```

# **Task - 3**

```python
texts = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for text in texts:
    tokens, clean = preprocess_text(text)

    print("Original Text:", text)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean)
    print("-" * 50)
```

# **Task - 4**

```python
def token_analytics(tokens):
    if not tokens:
        return 0, 0, 0

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(t) for t in tokens) / total

    return total, unique, round(avg_len, 2)
for text in texts:
    tokens, clean = preprocess_text(text)
    total, unique, avg_len = token_analytics(tokens)

    print("Original:", text)
    print("Tokens:", tokens)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Avg Token Length:", avg_len)
    print("-" * 50)
```

# **Task - 5**

```python
from collections import Counter

all_tokens = []

for text in texts:
    tokens, _ = preprocess_text(text)
    all_tokens.extend(tokens)

# Count frequency
counter = Counter(all_tokens)

# Top 10 frequent words
top_10 = counter.most_common(10)

# Least 5 frequent words
least_5 = counter.most_common()[-5:]

print("Top 10 Frequent Words:", top_10)
print("Least 5 Frequent Words:", least_5)
```

# **Task - 6**

```python
def full_pipeline(text_list):
    # Error handling
    if not isinstance(text_list, list):
        return {"tokens": [], "clean_sentences": []}

    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
result = full_pipeline(texts)

print("All Tokens:")
print(result["tokens"])

print("\nClean Sentences:")
print(result["clean_sentences"])
```
