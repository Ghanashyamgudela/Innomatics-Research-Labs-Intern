```python
def smart_parking_system(capacity, logs):
    current = 0
    peak = 0

    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT":
            current -= 1

        # Track peak usage
        if current > peak:
            peak = current

    print("Currently Parked Vehicles:", current)
    print("Peak Parking Usage:", peak)

    if current > capacity:
        print("Parking Status: ALERT! Capacity Exceeded")
    else:
        print("Parking Status: Available")


# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, vehicle_logs)
```

```python
def estimate_delivery_time(distance, traffic, weather):
    # Base delivery time
    base_time = distance * 5

    # Traffic delay
    if traffic == "Low":
        traffic_delay = 5
    elif traffic == "Medium":
        traffic_delay = 10
    elif traffic == "High":
        traffic_delay = 15
    else:
        traffic_delay = 0

    # Weather delay
    if weather == "Clear":
        weather_delay = 0
    elif weather == "Cloudy":
        weather_delay = 5
    elif weather == "Rainy":
        weather_delay = 10
    else:
        weather_delay = 0

    eta = base_time + traffic_delay + weather_delay

    print("Estimated Delivery Time:", eta, "minutes")


# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

estimate_delivery_time(distance, traffic, weather)
```

```python
def seat_occupancy_analyzer(total_seats, booked_seats):
    booked_count = len(booked_seats)

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
        print("Suggestion: Open additional shows")
    elif occupancy >= 70:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# Sample Input
total_seats = 200
booked_seats = [1]*150   # 150 seats booked

seat_occupancy_analyzer(total_seats, booked_seats)
```

```python
def classify_server_load(cpu_readings):
    total = 0

    # Loop to calculate total CPU usage
    for cpu in cpu_readings:
        total += cpu

    # Calculate average
    avg = total / len(cpu_readings)

    print("Average CPU Load:", int(avg), "%")

    # Classification
    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

```python
def monitor_resource_usage(resources):
    threshold = 8
    overused = []

    for resource, hours in resources.items():
        if hours > threshold:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resource_usage(resource_usage)
```

```python
def event_registration_controller(capacity, registrations):
    confirmed = 0
    waitlist = 0

    # Loop through registrations
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlist += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)

    if confirmed == capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


# Sample Input
capacity = 100
registrations = 105

event_registration_controller(capacity, registrations)
```
