```python

# Define template
template = "Explain {topic} for {audience} in a {tone} tone"

# Create PromptTemplate
prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template
)

# Function
def explain_topic(topic, audience, tone):
    return prompt.format(topic=topic, audience=audience, tone=tone)
```

```python
print(explain_topic("AI", "beginners", "friendly"))
print(explain_topic("Python", "kids", "fun"))
print(explain_topic("Deep Learning", "engineers", "technical"))
```

```python

# Teaching style
teaching_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Interview style
interview_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

# Storytelling style
story_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)
```

```python
def run_prompt_variations(topic):
    outputs = {
        "Teaching": teaching_prompt.format(topic=topic),
        "Interview": interview_prompt.format(topic=topic),
        "Storytelling": story_prompt.format(topic=topic)
    }
    return outputs
```

```python
results = run_prompt_variations("Machine Learning")

for style, output in results.items():
    print(f"{style}: {output}\n")
```

```python
from langchain_core.prompts import ChatPromptTemplate
def get_chat_prompt(role, topic):

    if role == "teacher":
        system_message = "You are a helpful teacher who explains concepts clearly step by step."

    elif role == "interviewer":
        system_message = "You are an interviewer who asks thoughtful and challenging questions."

    elif role == "motivator":
        system_message = "You are a motivational speaker who inspires and encourages learning."

    else:
        system_message = "You are a helpful assistant."

    # Create chat template
    chat_prompt = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", "Tell me about {topic}")
    ])

    return chat_prompt.format_messages(topic=topic)
```

```python
messages = get_chat_prompt("teacher", "Neural Networks")

for msg in messages:
    print(msg)
```

```python
def validate_inputs(audience, tone):

    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    # Validate audience
    if audience not in valid_audiences:
        print(f"Invalid audience '{audience}', defaulting to 'beginner'")
        audience = "beginner"

    # Validate tone
    if tone not in valid_tones:
        print(f"Invalid tone '{tone}', defaulting to 'formal'")
        tone = "formal"

    return audience, tone
```

```python
audience, tone = validate_inputs("kids", "friendly")

print(audience, tone)
```

```python
from langchain_core.prompts import PromptTemplate

# Base template
base_template = "Explain {topic} for {audience} in a {tone} {style} style"

prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone", "style"],
    template=base_template
)

# Validation
def validate_inputs(audience, tone):
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    if audience not in valid_audiences:
        audience = "beginner"

    if tone not in valid_tones:
        tone = "formal"

    return audience, tone
```

```python
def generate_prompt(topic, audience, tone, style):

    # Validate inputs
    audience, tone = validate_inputs(audience, tone)

    # Style control
    styles = ["teaching", "interview", "storytelling"]
    if style not in styles:
        style = "teaching"

    # Generate final prompt
    return prompt.format(
        topic=topic,
        audience=audience,
        tone=tone,
        style=style
    )
```

```python
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```
