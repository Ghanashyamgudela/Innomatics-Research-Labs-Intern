```python
# Ask the user to enter a username (input is taken as a string)
username = input("Username: ")

# Ask the user to enter a password
# int() is used because the password is expected to be a number
password = int(input("Password: "))

# Check if both username and password are correct
# If username is "admin" AND password is 1234 → Login Successful
# Otherwise → Invalid Credentials
print("Login Successful" if username == "admin" and password == 1234 else "Invalid Credentials")
```

```python
marks = [45, 78, 90, 33, 60]

l = [i for i in marks if i >= 50]   # students who passed
p = [i for i in marks if i < 50]    # students who failed

print("Total students Failed", len(l))
print("Total students Passed", len(p))
```

```python
# List of names with extra spaces and different letter cases
names = [" Alice ", "bob", " CHARLIE "]

# Empty list to store cleaned names
g = []

# Loop through each name in the list
for i in names:
    # strip() removes leading and trailing spaces
    # lower() converts the string to lowercase
    cleaned_name = i.strip().lower()

    # Add the cleaned name to the new list
    g.append(cleaned_name)

# Print the final cleaned list
print(g)
```

```python
# List of messages to analyze
messages = ["Hi", "Welcome to the platform", "OK"]

# Loop through each message in the list
for j in messages:

    # Calculate the length of the current message
    length = len(j)

    # Print the message
    print(f"Message: '{j}'")

    # Print the length of the message
    print(f"Length: {length}")

    # Check if the message is longer than 10 characters
    if length > 10:
        # Flag long messages
        print("Flag: Message is longer than 10 characters")

    # Print a blank line for better output readability
    print()
```

```python
# List of system log messages
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Variable to store the number of ERROR entries
count = 0

# Loop through each log message in the list
for log in logs:

    # Check if the current log message is "ERROR"
    if log == "ERROR":
        # Increase the count when an ERROR is found
        count += 1

# Print the total number of ERROR messages
print("Total ERROR entries:", count)
```
